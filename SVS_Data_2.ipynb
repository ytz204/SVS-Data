```python
import pandas as pd
```

```python
data = pd.read_csv("Admission_Predict_Ver1.1.csv")
```

# Descriptive Data

```python
data.describe()
```

# Mean GRE score of applicants who got admitted into a university with "5" rating

```python
uni_5 = data[data["University Rating"] >= 5]
```

```python
uni_5["GRE Score"].mean()
```

## We can take this average GRE score, 327/340, as the benchmark score for top tier grad school admission

# Find the mean TOEFL score for each University Rating

```python
data.groupby("GRE Score").mean()
```

```python
data.groupby("University Rating")["TOEFL Score"].mean()
```

## Conclusion; We can see a positive correlation between school tiers and toefl scores, and can use 113/120 as the mean benchmark sore for top tier schools

# Find the average research score per university rating

```python
data.groupby("University Rating")["Research"].mean()
```

## Conclusion: We can see that there is a positive correlation with average research score and university rating in grad school application, with the exception of lower tier schools
